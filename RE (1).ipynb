```python
# Model training

# Classes with 2 different folder
people = ['New_Monika', 'Other_eye']

DIR = r"E:\monik\RE_2"

features = []
labels = []

def create_train():
     for person in people:
         path = os.path.join(DIR, person)
         label = people.index(person)

         for img in os.listdir(path):
            img_path = os.path.join(path, img)
            img_array = cv.imread(img_path)

            if img_array is None :
                continue
            
            gray = cv.cvtColor(img_array, cv.COLOR_BGR2GRAY)
            features.append(gray)
            labels.append(label)

create_train()
print("Training data prepared!!")
```

```python
# Model training

# Classes with 2 different folder
people = ['New_Monika', 'Other_eye']

DIR = r"E:\monik\RE_2"

features = []
labels = []

def create_train():
     for person in people:
         path = os.path.join(DIR, person)
         label = people.index(person)

         for img in os.listdir(path):
            img_path = os.path.join(path, img)
            img_array = cv.imread(img_path)

            if img_array is None :
                continue
            
            gray = cv.cvtColor(img_array, cv.COLOR_BGR2GRAY)
            features.append(gray)
            labels.append(label)

create_train()
print("Training data prepared!!")

features = np.array(features, dtype = "object")
labels = np.array(labels)

# Create LBPH recognizer
eye_recognizer = cv.face.LBPHFaceRecognizer_create()

# Train
eye_recognizer.train(features, labels)

# Save trained model
eye_recognizer.save("eye_trained.yml")
np.save("eye_features.npy", features)
np.save("eye_labels.npy", labels)

print("Training completed and model saved!")
```
